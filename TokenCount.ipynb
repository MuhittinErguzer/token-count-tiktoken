# Token Analysis Tool for LLM Prompts

```python
# Importing libraries
import tiktoken
import re
```

```python
encoding = tiktoken.get_encoding("cl100k_base")
```

```python
# Get text input from user

print("Enter your text (press Enter twice to finish):")
lines = []
while True:
    line = input()
    if line == "" and len(lines) > 0 and lines[-1] == "":
        break
    lines.append(line)

text = "\n".join(lines).strip()
```

```python
# Find all words (sequences of letters, numbers, and underscores)
words = re.findall(r"\b\w+\b", text)
```

```python
# Encode full text
tokens = encoding.encode(text)
```

```python
# Print basic statistic
print(f"📝 FULL TEXT:")
print(text)
print("=" * 60)
print(f"📈 STATISTICS:")
print(f"   • Total Words:", len(words))
print(f"   • Total Tokens:", len(tokens))
print("=" * 60)
print(" --- WORDS WITH MULTIPLE TOKENS ---")
words_with_tokens = []
for word in words:
    word_tokens = encoding.encode(word)
    token_count = len(word_tokens)
    
    if token_count >= 2:
        split_parts = {encoding.decode([t]) for t in word_tokens}
        words_with_tokens.append({
            'word': word,
            'token_count': token_count,
            'split_parts': split_parts,
            'word_tokens': word_tokens
        })

# Sort by token_count in descending order (highest first)
words_with_tokens.sort(key=lambda x: x['token_count'], reverse=True)

# Print sorted results
for item in words_with_tokens:
    print(f"Word: '{item['word']}'")
    print(f"Token Count: {item['token_count']}")
    print(f"Token IDs: {item['word_tokens']}")
    print(f"Split Into: {item['split_parts']}")
    print("-" * 40)
```
