```python
# Import library
import pandas as pd
```

```python
# User Input Tokens
input_tokens = 187500000
output_tokens = 4890000000
cached_input_tokens = 20000000
```

```python
# Real model prices (USD per 1M tokens)
# Source OpenAI, Anthropic, Google official pages
models = {

	# OpenAI Models
	"gpt-5.5": {
		"provider": "OpenAI",
		"input": 5.00,
		"output": 30.00,
		"cached": 0.50
	},

	"gpt-5.4": {
		"provider": "OpenAI",
		"input": 2.50,
		"output": 15.00,
		"cached": 0.25
	},

	"gpt-5.5-mini": {
		"provider": "OpenAI",
		"input": 0.75,
		"output": 3.00,
		"cached": 0.075
	},
	
	"gpt-5.4-nano": {
		"provider": "OpenAI",
		"input": 0.05,
		"output": 0.40,
		"cached": 0.005
	},

	# Claude Models
	"Claude Opus 4.7": {
		"provider": "Anthropic",
		"input": 5.00,
		"output": 25.00,
		"cached": 0.50
	},

	"Claude Sonet 4.6": {
		"provider": "Anthropic",
		"input": 3.00,
		"output": 15.00,
		"cached": 0.30
	},
	"Claude Haiku 4.5": {
		"provider": "Anthropic",
		"input": 1.00,
		"output": 5.00,
		"cached": 0.10
	},

	#Gemini Models
	"Gemini 3.5 Flash": {
		"provider": "Google",
		"input": 1.50,
		"output": 9.00,
		"cached": 0.15
	},
	"Gemini 3 Flash": {
		"provider": "Google",
		"input": 0.50,
		"output": 3.00,
		"cached": 0.05
	},
	"Gemini 3.1 Flash-Lite": {
		"provider": "Google",
		"input": 0.25,
		"output": 1.50,
		"cached": 0.025
	},
}
```

```python
# Cost Calculation Function
def calculate_cost(input_tokens, output_tokens, cached_input_tokens, price):
	input_cost = (
		input_tokens / 1000000
	) * price["input"]
				
	output_cost = (
		output_tokens / 1000000
	) * price["output"]

	cached_cost = (
		cached_input_tokens / 1000000
	) * price["cached"]

	total_cost = (
		input_cost +
		output_cost +
		cached_cost
	)

	return (
		input_cost,
		output_cost,
		cached_cost,
		total_cost
	)
```

```python
# Build comparasion Table
rows = []

for model_name, price in models.items():
	input_cost, output_cost, cached_cost, total_cost = calculate_cost(
		input_tokens,
		output_tokens,
		cached_input_tokens,
		price
	)

	rows.append({
		"Provider":
			price["provider"],				
		"Model":
			model_name,
		"Input Cost $":
			round(input_cost, 2),	
		"Output Cost $":
			round(output_cost, 2),
		"Cached Input Cost $":
			round(cached_cost, 2),
		"Total Cost $":
			round(total_cost, 2)
	})
```

```python
# Create a Dataframe
df = pd.DataFrame(rows)

# Sort from cheapest to most expensive
df = df.sort_values(
	by="Total Cost $"
).reset_index(drop=True)
```

```python
# Display Final Table
print("\n======== LLM Cost Comparasion ==============\n")
print(df.to_string(index=False))

# Show cheapest model
cheapest = df.iloc[0]
print("\n=============================\n")
print("Cheapest Model")
print("\n=============================\n")
print(
	f"{cheapest['Provider']} - "
	f"{cheapest['Model']} = "
	f"{cheapest['Total Cost $']} - "
)
```
